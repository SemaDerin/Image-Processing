```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
foto= cv2.imread("negatif.jpg",0)
```

```python
def fotograf_negatifi(foto):
    L= np.max(foto)
    negatif_foto= L-foto
    return negatif_foto
```

```python
negatif_foto=fotograf_negatifi(foto)
```

```python
#yanyaya
yan_yana= np.hstack((foto,negatif_foto))
```

```python
plt.imshow(yan_yana, cmap="gray")
plt.show()
```
